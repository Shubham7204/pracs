```python
# Required Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Jupyter Magic Command
%matplotlib inline

# Helper function to display images using matplotlib
def show_img(img, title='Image', cmap_type=None):
    plt.figure(figsize=(5, 5))
    if len(img.shape) == 2:  # grayscale
        plt.imshow(img, cmap=cmap_type if cmap_type else 'gray')
    else:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load images
img1 = cv2.imread('C:\\Users\\HP\\Downloads\\Standard Images\\football.jpg')         # replace with your actual path
img2 = cv2.imread('C:\\Users\\HP\\Downloads\\Standard Images\\opencv-logo.png')      # replace with your actual path

# Show images
show_img(img1, "Image 1")
show_img(img2, "Image 2")

# Grayscale
gray = cv2.imread('C:\\Users\\HP\\Downloads\\Standard Images\\football.jpg', cv2.IMREAD_GRAYSCALE)
show_img(gray, "Grayscale")

# Cropping
crop = img1[0:200, 0:200]
show_img(crop, "Cropped Image")

# Adding image to itself
add_img = cv2.add(img1, img1)
show_img(add_img, "Image Added to Itself")

# Image Properties
print("Image 1 properties:")
print(f"Type: {type(img1)}")
print(f"Height: {img1.shape[0]}")
print(f"Width: {img1.shape[1]}")
print(f"Shape: {img1.shape}")
print(f"Dimensions: {img1.ndim}")

print("\nImage 2 properties:")
print(f"Type: {type(img2)}")
print(f"Height: {img2.shape[0]}")
print(f"Width: {img2.shape[1]}")
print(f"Shape: {img2.shape}")
print(f"Dimensions: {img2.ndim}")

# Split BGR Channels
b, g, r = cv2.split(img1)

# Display each channel
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(g, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(r, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

plt.show()

# Resize img2 to match img1
img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Arithmetic Operations
show_img(cv2.add(img1, img2_resized), "Addition")
show_img(cv2.subtract(img1, img2_resized), "Subtraction")
show_img(cv2.multiply(img1, img2_resized), "Multiplication")
show_img(cv2.divide(img1, img2_resized), "Division")

# Bitwise Operations
show_img(cv2.bitwise_and(img1, img2_resized), "Bitwise AND")
show_img(cv2.bitwise_or(img1, img2_resized), "Bitwise OR")
show_img(cv2.bitwise_xor(img1, img2_resized), "Bitwise XOR")
show_img(cv2.bitwise_not(img1), "Bitwise NOT")

# Reflection
img = cv2.imread('C:\\Users\\HP\\Downloads\\Standard Images\\messi.jpg')  # replace with your actual path
show_img(img, "Original Messi")

rows, cols, _ = img.shape

# Horizontal reflection using warpPerspective
M = np.array([[-1, 0, cols], [0, 1, 0], [0, 0, 1]], dtype=np.float32)
reflected_img = cv2.warpPerspective(img, M, (cols, rows))
show_img(reflected_img, "Reflected Image")
```
