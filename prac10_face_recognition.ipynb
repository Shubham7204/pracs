```python
import cv2
import matplotlib.pyplot as plt

# Helper function to show frame using matplotlib
def show_frame(img, title="Frame"):
    plt.figure(figsize=(8, 6))
    if len(img.shape) == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load video file (update this path as needed)
video_path = 'C:\\Users\\HP\\Downloads\\openai.mp4'  # Adjust path for Jupyter environment
cap = cv2.VideoCapture(video_path)

print("Face Detection using Haar Cascade")

# Load Haar Cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

frame_count = 0

while frame_count < 10:
    ret, frame = cap.read()
    if not ret:
        print("Can't read frame or video ended.")
        break

    # Convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    # Draw rectangles around detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # Show the frame every time (or use `if frame_count % n == 0` to reduce output)
    show_frame(frame, f"Frame {frame_count + 1}")

    frame_count += 1

# Release video capture
cap.release()
print(f"Processed {frame_count} frames.")
```
